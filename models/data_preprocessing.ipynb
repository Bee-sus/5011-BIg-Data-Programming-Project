```python
data = data.drop(columns=["id", "first_name", "last_name", "email"])
```

```python
data['part_time_job'] = data['part_time_job'].astype(int)
```

```python
data['extracurricular_activities'] = data['extracurricular_activities'].astype(int)
```

```python
data
```

```python
data['gender'] = data['gender'].astype(str).str.strip().str.lower()
data['gender'] = data['gender'].map({'male': 0, 'female': 1})
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, StandardScaler

encoder = OneHotEncoder(sparse_output=False)
career_encoded = encoder.fit_transform(data[['career_aspiration']])
career_data = pd.DataFrame(career_encoded, columns=encoder.get_feature_names_out(['career_aspiration']))
data = pd.concat([data.drop(columns=['career_aspiration']), career_data], axis=1)
```

```python
data
```

```python
score_cols = ['math_score', 'history_score', 'physics_score','chemistry_score', 'biology_score', 'english_score', 'geography_score']

data['average_score'] = data[score_cols].mean(axis=1)
data['science_avg'] = data[['math_score', 'physics_score', 'chemistry_score', 'biology_score']].mean(axis=1)
data['humanities_avg'] = data[['history_score', 'english_score', 'geography_score']].mean(axis=1)
```

```python
data
```

```python
data.to_csv('student-scores-preprocessed.csv', index=False)
```

```python
import pandas as pd
import numpy as np

# Load dataset
data = pd.read_csv("student-scores-preprocessed.csv")
data
```

```python
# --- Feature 1: Engagement Score ---
data["engagement_score"] = (data["weekly_self_study_hours"]+ 5 * data["extracurricular_activities"]- 0.5 * data["absence_days"])
```

```python
data
```

```python
# --- Feature 2: Performance Consistency ---
subject_cols = ["math_score", "history_score", "physics_score", "chemistry_score","biology_score", "english_score", "geography_score"]

data["performance_consistency"] = 100 - data[subject_cols].std(axis=1)

# --- Optional: Clip consistency to [0,100] for safety ---
data["performance_consistency"] = data["performance_consistency"].clip(0, 100)

# --- Display result ---
print(data[["engagement_score", "performance_consistency"]].head())
```

```python
data
```

```python
data.to_csv("student-scores-enhanced.csv", index=False)
print("Feature engineering completed. File saved as 'student-scores-enhanced.csv'")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load enhanced dataset
data = pd.read_csv("student-scores-enhanced.csv")

# Basic setup
sns.set(style="whitegrid", palette="pastel")
plt.rcParams["figure.figsize"] = (8, 5)
```

```python
sns.histplot(data["average_score"], kde=True, bins=20, color="skyblue")
plt.title("Distribution of Average Scores")
plt.xlabel("Average Score")
plt.ylabel("Number of Students")
plt.show()
```

```python
sns.scatterplot(data=data, x="engagement_score", y="average_score", hue="gender", alpha=0.7)
plt.title("Engagement Score vs. Average Score by Gender")
plt.xlabel("Engagement Score")
plt.ylabel("Average Score")
plt.show()
```

```python
sns.scatterplot(data=data, x="performance_consistency", y="average_score", color="purple", alpha=0.6)
plt.title("Performance Consistency vs. Average Score")
plt.xlabel("Performance Consistency (100 - Std Dev of Subject Scores)")
plt.ylabel("Average Score")
plt.show()
```

```python
sns.boxplot(data=data, x="part_time_job", y="average_score", hue="gender")
plt.title("Average Score by Gender and Part-Time Job Status")
plt.xlabel("Part-Time Job (0 = No, 1 = Yes)")
plt.ylabel("Average Score")
plt.show()
```

```python
corr = data[["average_score", "engagement_score", "performance_consistency","weekly_self_study_hours", "absence_days", "science_avg", "humanities_avg"]].corr()

sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Key Features")
plt.show()
```

```python
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt

kmeans = KMeans(n_clusters=3, random_state=42)
data["performance_cluster"] = kmeans.fit_predict(
    data[["average_score", "weekly_self_study_hours", "absence_days"]]
)

sns.scatterplot(
    data=data,
    x="weekly_self_study_hours",
    y="average_score",
    hue="performance_cluster",
    palette="viridis",
    alpha=0.8
)
plt.title("Student Clusters by Performance")
plt.xlabel("Weekly Self Study Hours")
plt.ylabel("Average Score")
plt.show()

```
